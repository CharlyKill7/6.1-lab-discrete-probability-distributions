## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""

# De momento, se me ocurre esta forma de adaptar el espacio muestral.
espacio_muestral = ['apple', 'apple', 'apple', 'orange', 'orange']

# Saco 100 frutas de la caja

import random

sacadas = random.choices(espacio_muestral, k=100)

sacadas[:10]
````

````python
# frecuancia (probabilidad)

frec_apple = sacadas.count('apple')/len(sacadas)
frec_apple
````

````python
frec_orange = sacadas.count('orange')/len(sacadas)
frec_orange
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
import pylab as plt

from scipy.stats import binom
````

````python
# ¿Cual es la prob de sacar cinco apples si saco 20 frutas?

#binom.pmf(k, n, p)  # k=nº de exitos, n=nº de repeticiones, p=prob, pmf=probability mass function


binom.pmf(5, 20, 0.6)
````

````python
x=[i for i in range(20)]                  # nº de exitos

y=[binom.pmf(x, 20, 0.6) for x in x]      # su probabilidad

plt.plot(x, y);
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
# Aquí estoy probando cosas.

# cdf=cumulative density function

x=[i for i in range(20)]                  # nº de exitos
y=[binom.cdf(x, 20, 0.6) for x in x]      # su probabilidad

plt.plot(x, y);
````

````python
area_total = binom.cdf(20, 20, 0.6)
area_total
````

````python
area_app_4ol = binom.cdf(4, 20, 0.6)
area_app_4ol
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math
from scipy.stats import poisson
````

````python
# poisson.pmf(k, lambda), poisson.pmf(evento, media)

# media de goles es 2.5, ¿prob de marcar 5?

poisson.pmf(5, 2.3).round(3)
````

Si X ~ Poisson(λ) entonces la función de probabilidad es

P(X=k) = (e^(-λ)λ^k) / k!

donde k = 0, 1, 2, … es el número de ocurrencias del evento o fenómeno.

El parámetro λ > 0 representa el número de veces que se espera que ocurra dicho fenómeno durante un intervalo dado. Por ejemplo, si el suceso estudiado tiene lugar en promedio 4 veces por minuto y estamos interesados en la probabilidad de que ocurra k veces dentro de un intervalo de 10 minutos, usaremos un modelo de distribución de Poisson con λ = 10×4 = 40.


````python
num = (math.e ** (-2.3)) * (2.3**5)
````

````python
prob = num / math.factorial(5)
prob
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
x=[i for i in range(10)]    # numero de  goles

y=[poisson.pmf(x, 2.3) for x in x] 

plt.plot(x, y);

plt.plot(x, y)
plt.xlabel('Nº de goles') 
plt.ylabel('Probabilidad') 
plt.title('Cumulative Mass Function')
plt.show;
````
